## Summarization test

````python
text = """
The Transformer is a neural network architecture based solely on attention mechanisms,
dispensing with recurrence and convolutions entirely. Experiments on two machine
translation tasks show these models to be superior in quality while being more
parallelizable and requiring significantly less time to train.
"""

summary = llm.generate(
    f"Summarize the following text in 2 bullet points:\n\n{text}",
    system_prompt="You are an expert at summarizing technical text.",
    max_tokens=200,
)

print(summary)
````

## Question answering test

````python
context = """
Attention Is All You Need introduces the Transformer architecture.
Self-attention allows the model to weigh the importance of different words
in a sequence when encoding a token.
"""

question = "What does self-attention allow the model to do?"

qa_response = llm.generate(
    f"Context:\n{context}\n\nQuestion: {question}\n\nAnswer based only on the context:",
    system_prompt="Answer only using the provided context. If unsure, say you don't know.",
    max_tokens=150,
)

print(f"Question: {question}")
print(f"Answer: {qa_response}")
````
